```python
import pandas as pd
import numpy as np
import csv
import os
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
from sklearn import linear_model
%matplotlib inline
```

# Load Philadelphia Crime and Housing Price Data

```python
data = pd.read_csv('Philadelphia_Crime_Rate_noNA.csv', sep = ',')
```

```python
print data.head()
```

# Explore the Data

```python
data.plot(kind = 'scatter', x = 'CrimeRate', y = 'HousePrice', color = 'red')
```

# Fit the Regression Model

```python
price_crime = data[['CrimeRate','HousePrice']]
np_price_crime = np.matrix(price_crime)
lr = linear_model.LinearRegression()
x,y = np_price_crime[:,0], np_price_crime[:,1]
lr.fit(x, y)
```

```python
plt.plot(data['CrimeRate'],data['HousePrice'],'r.',
        data['CrimeRate'],lr.predict(x),'b-')
```

# Remove High Crime Rate Outlier and Refit the Regression

```python
data_noHCR = data[data['MilesPhila'] != 0.0]
```

```python
data_noHCR.plot(kind = 'scatter', x = 'CrimeRate', y = 'HousePrice', color = 'red')
```

```python
np_data_noHCR = np.matrix(data_noHCR[['CrimeRate','HousePrice']])
lr1 = linear_model.LinearRegression()
x1, y1 = np_data_noHCR[:,0], np_data_noHCR[:,1]
lr1.fit(x1, y1)
```

```python
plt.plot(data_noHCR['CrimeRate'], data_noHCR['HousePrice'], 'r.',
         data_noHCR['CrimeRate'], lr1.predict(x1), 'b-')
```

# Compare Coefficient and Intercept

```python
print "Coefficient (original data): {}" .format(lr.coef_)
print "Coefficient (no high crime rate): {}" .format(lr1.coef_)
print "--------------------------------------------------------"
print "Intercept (original data): {}" .format(lr.intercept_)
print "Intercept (no high crime rate): {}" .format(lr1.intercept_)
```

# Remove High House Price and Refit the Regression

```python
data_noHHP = data_noHCR[data_noHCR['HousePrice'] < 350000]
```

```python
data_noHHP.plot(kind = 'scatter', x = 'CrimeRate', y = 'HousePrice', color = 'red')
```

```python
np_data_noHHP = np.matrix(data_noHHP[['CrimeRate', 'HousePrice']])
lr2 = linear_model.LinearRegression()
x2, y2 = np_data_noHHP[:,0], np_data_noHHP[:,1]
lr2.fit(x2, y2)
```

```python
plt.plot(data_noHHP['CrimeRate'], data_noHHP['HousePrice'], 'r.',
         data_noHHP['CrimeRate'], lr2.predict(x2), 'b-')
```

# Compare Coefficient and Intercept

```python
print "Coefficient (no high crime rate): {}" .format(lr1.coef_)
print "Coefficient (no high house price): {}" .format(lr2.coef_)
print "--------------------------------------------------------"
print "Intercept (no high crime rate): {}" .format(lr1.intercept_)
print "Intercept (no high house price): {}" .format(lr2.intercept_)
```

#### We see that the the data point with the highest crime rate is very far from the rest of the data and thus has dramatic impact on the regression model compared to those with higher housing price.
